```python
'''
Import libraries
'''
# Import here useful libraries
import numpy as np
from scipy import stats
import scipy
import warnings
from sympy import symbols
from sympy import integrate
import matplotlib.pyplot as plt
warnings.simplefilter('ignore', DeprecationWarning)
```

```python
from scipy.stats import norm

# Parameters
x = 12 # value to look for
mu = 12 # mean
sigma = 14 # standard deviation

print("Mean: ", norm.mean(loc = mu, scale = sigma))
print("Variance: ", norm.var(loc = mu, scale = sigma)) 
print("Probability mass function: ", norm.pdf(x, loc = mu, scale = sigma))
print("Cumulative distribution function: ", norm.cdf(x, loc = mu, scale = sigma))
print("Survival function (1-cdf): ", norm.sf(x, loc = mu, scale = sigma))
```

```python
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
ax=stats.norm.pdf(x, mu, sigma)
plt.plot(x, ax)
plt.xticks(np.arange(mu - 3*sigma,mu + 3*sigma,5))
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.title("HW#5, P3")
plt.xlabel("Temperature, T")
plt.ylabel("fT(t)")
#ax.set_ylim([0, 0.03])
plt.ylim(bottom=0, top=0.03)
plt.show()
```
