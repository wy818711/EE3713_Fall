```python
'''
Import libraries
'''
# Import here useful libraries
import numpy as np
import math as math
from scipy import stats
import scipy
import warnings
from sympy import symbols
from sympy import integrate
import matplotlib.pyplot as plt
warnings.simplefilter('ignore', DeprecationWarning)
```

```python
# rv_discrete is used to define the discrete RV distribution. Please read the following link for rv_discrete which is from scipy library.
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_discrete.html
```

```python
#Example 1ï¼š Expected value practice
from scipy.stats import rv_discrete
x = [10, 20, 30] # Give the random variable X
p = [0.2, 0.3, 0.5] # The probability of correspond to X
distribution = rv_discrete(values=(x, p)) # Define the distribution as discrete RV
print("Expected value: ", distribution.expect())
print("Standard deviation: ", distribution.std())
print("Variance: ", distribution.var())
```

```python
#Example 2: CDF and PMF plotting
#You are given the CDF (FX(x)) that 0 for x <-1, 0.2 for -1=<x<0, 0.7 for 0=<x<1, 1 for x>=1.
#You are asked to plot CDF and PMF. You are also asked to provide the Expected value, Variance, and Standard deviation.
#So, your CDF should be CDF=(0, 0.2, 0.7, 1) with corresponding x as (-1,0,1).
CDF=(0, 0, 0.2, 0.7, 1) # 0 is added at the beginning to align the plot and make the plot looks better.
x=(-2,-1,0,1,2) # -2 and 2 are added at the beginning and end to align the plot and make the plot looks better.
fig, ax = plt.subplots(1, 1)
plt.step(x,CDF,where='pre' )
plt.title("CDF")
plt.xlabel("x")
plt.ylabel("FX(x)")
plt.ylim(bottom=-0.1, top=1.1)
plt.xlim(left=-2, right=2)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
# Calculate PMF from CDF. 
CDF=(0, 0.2, 0.7, 1)
n = len(CDF)
output = []
for i in range(len(CDF) - 1):
  result = CDF[i] - CDF[i + 1]
  output.append(-result)
print("PMF=", np.around(output,1))
```

```python
#Example 2 (continuous): PMF plotting
from scipy import stats
xk = (-1, 0, 1) # Give x
pk = (0.2, 0.5, 0.3) # Corresponding PMF
custm = stats.rv_discrete(name='custm', values=(xk, pk))
Ex=np.sum(np.dot(xk,pk)) # Find out the expected value from PMF E[X]=sum of (xk*)*(pk)
Var=np.sum(np.dot((np.square(xk)),pk))-pow(Ex,2)
std=math.sqrt(Var)
print("Expected value:",round(Ex,1)) # Print out the expected value from PMF.
print("Variance:",round(Var,2)) # Print out the variance value from PMF.
print("Standard deviation:",round(std,2)) # Print out the std from PMF.
import matplotlib.pyplot as plt # Call figure
fig, ax = plt.subplots(1, 1)
ax.plot(xk, custm.pmf(xk), 'ro', ms=12, mec='r')
ax.vlines(xk, 0, custm.pmf(xk), colors='r', lw=2)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.title("PMF")
plt.xlabel("x")
plt.ylabel("pX(x)")
plt.ylim(bottom=0, top=0.6)
plt.show()
```
